```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# データの準備
x1 = np.linspace(-5, 5, 101)
y1 = np.sin(x1)
x2 = np.random.rand(100)
y2 = x2 * np.random.rand(100)
```

```python
# 折れ線グラフの作成
# しっかりオブジェクトを作りましょう
fig, ax = plt.subplots()
ax.set_title("y = sinx")
ax.set_xlabel("X")
ax.plot(x1, y1)
```

```python
# 散布図の描画
fig, ax = plt.subplots()
ax.set_title("scatter")
ax.set_xlabel("x2")
ax.set_ylabel("y2")
ax.scatter(x2, y2)
```

```python
# ヒストグラムの描画
plt.hist(x2, bins=20)
```

```python
# 箱ひげ図の描画
plt.boxplot(y2)
```

```python
# 棒グラフの描画
plt.bar(x2, y2)
```
